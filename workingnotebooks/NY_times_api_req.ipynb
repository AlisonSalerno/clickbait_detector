```python
from pynytimes import NYTAPI
```

```python
nyt = NYTAPI("yourkey")
```

```python
import datetime

data = nyt.archive_metadata(
    date = datetime.datetime(2020, 6, 1)
)
```

```python
data[0]['headline']['main']
```

```python
headlines_nyt=[]
for i in range(len(data)):
    headlines_nyt.append((data[i]['headline']['main'],data[i]['pub_date']))
```

```python
headlines_nyt
```

```python
import pandas as pd
ny_times_datafinal = pd.DataFrame(headlines_nyt,columns=['text','date'])
```

```python
ny_times_datafinal.to_csv('nytimes_data_final.csv')
```
